### Reshaping and Pivot Tables

```python
import pandas as pd
```

```python
#Pivoting data frame objects
DX = pd.read_csv('Pivot.csv')
DX
```

```python
DX.pivot(index="Person", columns="Day")
```

```python
X3 = DX.pivot(index="Day", columns="Person", values="DistanceRan")
X3
```

```python
isinstance(X3, pd.DataFrame)
```

```python
X3['Mark']
```
